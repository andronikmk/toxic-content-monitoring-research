```python
import basilica
import numpy as np
import pandas as pd
```

### Disclaimer:
Embedded the first 10,000 rows of the toxic-clean-train.csv, then decided to do the entire dataset. Note, it is 1 row short so "basilica_y.csv" will need to be used when implementing it in a model.

```python
# dataframe was cleaned (url's, usernames, unicode, newline and new characters removed):
df = pd.read_csv('toxic-train-clean.csv')
# y columns are levels of toxicity ('toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate')
y = df.iloc[:, 2:8][:10000]
x = df['comment_text'][:10000]
```

```python
class Embedder():
    """
    Creates connection and Basilica embeddings for text
    """
    def __init__(self):
        self.API_KEY='API_KEY'
        
    def get_vectors(self, text):
        with basilica.Connection(self.API_KEY) as c:
            embedded = (c.embed_sentence(text))
        return embedded
be = Embedder()
```

```python
# runs through dataframe creating vectors for each row:
i = 0
l = len(x)
def vectorize_all(text):
    global i
    global l
    i+=1
    if i % 100 == 0:
        print(i, "/", l)
    return be.get_vectors(text)
```

```python
# puts embeddings in proper format to run through models:
embeddings = np.stack(x.apply(vectorize_all).values)
```

```python
len(embeddings)
```

```python
np.savetxt('10k_basilica_toxic_clean.out', embeddings, delimiter=',')
```

```python
# remaining dataset to be embedded (this is where indexing oversight occurs, index 10000 is not accounted for)
rest = df['comment_text'][10001:]
resty = df.iloc[:, 2:8][10001:]

rest.shape, resty.shape
```

```python
# runs through dataframe creating vectors for each row:
rest_i = 0
rest_l = len(rest)
def vectorize_all(text):
    global rest_i
    global rest_l
    rest_i+=1
    if rest_i % 100 == 0:
        print(rest_i, "/", rest_l)
    return be.get_vectors(text)
```

```python
embeddings_rest = np.stack(rest.apply(vectorize_all).values)
```

```python
np.savetxt('rest_basilica_toxic_clean.out', embeddings_rest, delimiter=',')
```

```python
# combining the first 10k with the remaininng rows of database:
basilica_embeddings = np.concatenate((embeddings, embeddings_rest), axis=0)
```

```python
np.savetxt('basilica_toxic_clean.out', basilica_embeddings, delimiter=',')
```

```python
# combining the y classes for the first 10k with the remaining rows (need to do this because it is 1 row short)
basilica_y = pd.concat((y, resty), axis=0)
```

```python
basilica_y.to_csv('basilica_y.csv', index=False)
```
